```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Chrips 1998-2008.csv')
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python


# Extract year and month
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month

# Group by year and month, then calculate the average rainfall
monthly_average = df.groupby(['year', 'month'])['ave_rainfall'].sum().reset_index()

monthly_average
```

```python
yearly_average = df.groupby('year')['ave_rainfall'].sum().reset_index()
yearly_average
```

```python
yearly_average.to_csv('chrips_yearly_rainfall_data.csv', index=False)
```
